Изучение и обработка выбросов


Учитывая приведенные выше результаты, столбцы, содержащие выбросы, — это 'Tomatometer rating', 'Year of Release', 'IMDB Rating', и 'Audience rating'. Создадим DataFrame без этих выбросов. Выбросы можно идентифицировать путем определения нижнего/верхнего предела нормального диапазона значений.

```python
#выбросы
IMDB = data.query('Award == ["Winner","Nominee"]')
IMDB.boxplot(by ='Award', column ='IMDB Rating', grid = True)
```

```python
# Определить по приведенной выше гистограмме столбцы, содержащие выбросы: Tomatometer Rating,Year of Release, IMDB Rating и Audience Rating.
# Создать Dataframe на основе столбцов с выбросами
outlier = data[['Tomatometer Rating','Audience Rating','Year of Release', 'IMDB Rating']]
outlier
```

```python
# Определить нижние пределы для выбросов
Q1 = outlier.quantile(0.25)
Q1
```

```python
# Определить верхние пределы выбросов
Q3 = outlier.quantile(0.75)
Q3
```

```python
# IQR-оценки
IQR = Q3-Q1
# Отфильтрованные данные, из которых удалены выбросы
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
# Копирование индексов
index_list = list(data_filtered.index.values)
```

```python
# Хранить данные без выбросов в отдельном DataFrame.
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

```python
# создать гистограмму с отфильтрованным выбросом
filtered_outlier = data_filtered[['Tomatometer Rating','Audience Rating', 'IMDB Rating','Year of Release']]

for col in filtered_outlier:
    data_filtered.hist(col, bins=100)
```

Как мы видим, рейтинг Аудитории и Томатометра коррелируются и близки по значениям. IMDB рейтинг же находится в основом на уровне 7.6, что ниже чем рейтинг Аудитории и Томатометра.
