```python
df['color'] = df['color'].str.lower() 
unique_colors = df['color'].unique()
unique_colors
```

```python
import re #Честно был без понятия как корректно изменить данные, так что подсмотрел и украл немного кода ( тут регулярное выражение)
pattern = r'(color|black and white|colour|unknown)'
df['color'] = df['color'].str.extract(pattern, flags=re.IGNORECASE)
print(df['color'].unique())
```

```python
for row in ['color']:
    df[row] = df[row].fillna('unknown')
print(df['color'].unique())
```

```python
df.groupby ('color')['duration'].count()
```

```python
df['color'] = df['color'].replace({
    ' Black and White': 'Black and White',
    'colour': 'color'
})
```

```python
df.groupby ('color')['duration'].count()
```

Наконец то сама гипотеза
